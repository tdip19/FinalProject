```python
import pandas as pd
```

```python
# Import all the data

sp500 = pd.read_excel('inputs/SP500_return_quarterly.xlsx')
sp500
```

```python
gdp = pd.read_excel('inputs/GDP_by_quarter.xlsx')
gdp = gdp.drop(['Table 1.1.1. Percent Change From Preceding Period in Real Gross Domestic Product'], axis=1)
gdp = gdp.drop(range(10))
gdp.columns = gdp.iloc[0]
gdp = gdp.iloc[1:]
gdp = gdp.reset_index(drop=True)
gdp
```

```python
cpi = pd.read_excel('inputs/CPI_by_quarter.xls')
# cpi = cpi.drop(['FRED Graph Observations'], axis=1)
cpi = cpi.drop(range(9))
cpi.columns = cpi.iloc[0]
cpi = cpi.iloc[1:]
cpi = cpi.reset_index(drop=True)
# cpi = cpi.rename(columns={cpi.columns[0]: 'Period', cpi.columns[1]: 'Label'}, level=0)
cpi

# note: this dataframe has data for Q1 2023 whereas most others don't
```

```python
constr_unmeployment = pd.read_excel('inputs/construction_unemployment.xlsx')
# constr_unmeployment = constr_unmeployment.drop(['FRED Graph Observations'], axis=1)
constr_unmeployment = constr_unmeployment.drop(range(18))
constr_unmeployment.columns = constr_unmeployment.iloc[0]
constr_unmeployment = constr_unmeployment.iloc[1:]
constr_unmeployment = constr_unmeployment.reset_index(drop=True)
constr_unmeployment

# note: this dataframe also has data for Q1 2023 whereas most others don't
```

```python
manuf_unemployment = pd.read_excel('inputs/manufacturing_unemployment.xlsx')
# gdp = gdp.drop(['Table 1.1.1. Percent Change From Preceding Period in Real Gross Domestic Product'], axis=1)
manuf_unemployment = manuf_unemployment.drop(range(18))
manuf_unemployment.columns = manuf_unemployment.iloc[0]
manuf_unemployment = manuf_unemployment.iloc[1:]
manuf_unemployment = manuf_unemployment.reset_index(drop=True)
manuf_unemployment
```

```python
finance_unemployment = pd.read_excel('inputs/finance_unemployment.xlsx')
# gdp = gdp.drop(['Table 1.1.1. Percent Change From Preceding Period in Real Gross Domestic Product'], axis=1)
finance_unemployment = finance_unemployment.drop(range(18))
finance_unemployment.columns = finance_unemployment.iloc[0]
finance_unemployment = finance_unemployment.iloc[1:]
finance_unemployment = finance_unemployment.reset_index(drop=True)
finance_unemployment

# note: this dataframe also has data for Q1 2023 whereas most others don't
```

```python
# rename 'Observation Value' to whatever thee statistic is
sp500 = sp500.rename(columns={'Observation value': 'SP500_Return'})
gdp = gdp.rename(columns={'Observation Value': 'Percent_Change_in_GDP'})
cpi = cpi.rename(columns={'Observation Value': 'CPI'})
constr_unmeployment = constr_unmeployment.rename(columns={'Observation Value': 'Construction_Unemployment_Rate'})
manuf_unemployment = manuf_unemployment.rename(columns={'Observation Value': 'Manufacturing_Unemployment_Rate'})
finance_unemployment = finance_unemployment.rename(columns={'Observation Value': 'Finance_Unemployment_Rate'})
```

```python
# drop last row of those with a 2023 Q1 observation to preparee for merge

constr_unmeployment = constr_unmeployment.drop(constr_unmeployment.index[-1])
cpi = cpi.drop(cpi.index[-1])
finance_unemployment = finance_unemployment.drop(finance_unemployment.index[-1])
manuf_unemployment = manuf_unemployment.drop(manuf_unemployment.index[-1])
cpi
```

```python
merged_data = sp500.merge(gdp,
                          how='inner',
                          on=['Year', 'Period', 'Label'],
                          validate = '1:1')

merged_data = merged_data.merge(cpi,
                          how='inner',
                          on='Label',
                          validate = '1:1')

merged_data = merged_data.merge(constr_unmeployment,
                          how='inner',
                          on='Label',
                          validate = '1:1')

merged_data = merged_data.drop(['Year_x', 'Period_x', 'Year_y', 'Period_y'], axis=1)

merged_data = merged_data.merge(manuf_unemployment,
                          how='inner',
                          on='Label',
                          validate = '1:1')

merged_data = merged_data.merge(finance_unemployment,
                          how='inner',
                          on='Label',
                          validate = '1:1')

merged_data = merged_data.drop(['Year_x', 'Period_x', 'Year_y', 'Period_y'], axis=1)

# move year and period column to front
year_col = merged_data.pop('Year')
period_col = merged_data.pop('Period')
merged_data.insert(0, 'Period', period_col)
merged_data.insert(0, 'Year', year_col)
```
